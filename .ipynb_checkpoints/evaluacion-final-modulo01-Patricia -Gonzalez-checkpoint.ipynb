```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []  # lista vacía para almacenar datos desde cero
        self.clientes = {}  # diccionario vacío
        self.ventas_totales = 0.0 # para que acepte numeros decimales
    
    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre: #["nombre"] se refiere al producto que se quiere agregar---el input de usuario
                producto["cantidad"] += cantidad #suma cantidad actual con cantidad actualizada
                print(f"{nombre} se ha actualizado")
                return 
        self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
        print(f"{nombre} se ha agregado al inventario")
            
    
    def ver_inventario(self):
        for producto in self.inventario:
            print(f'Nombre: {producto["nombre"]}, Precio: {producto["precio"]} €, Disponible: {producto["cantidad"]}')
    
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                print(f'Nombre: {producto["nombre"]}, Precio: {producto["precio"]} €, Disponible: {producto["cantidad"]}')
                return producto
        print(f"{nombre} no está en el inventario")
        return None
    
    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return f'Inventario {nombre} actualizado a {producto["cantidad"]}'
        return f"'{nombre}' no existe en el inventario."
    
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                self.inventario.remove(producto)
                return f"{nombre} ha sido eliminado del inventario."
        return f"{nombre} no existe en el inventario."
    
    def calcular_valor_inventario(self):
        valor_total = sum(producto["precio"] * producto["cantidad"] for producto in self.inventario)
        return f"El valor total del inventario es {valor_total} €."
    
    def realizar_compra(self):
        self.ver_inventario()  # no funciona
        carrito = []  # variable carrito para poder iniciar 
        while True:
            nombre_producto = input("Ingresa el producto que deseas comprar. Escribe ('salir') para finalizar la compra").lower()
            if nombre_producto == "salir":
                break # DETIENE el bucle. Si no, el codigo continúa.
            producto = self.buscar_producto(nombre_producto) #producto ingresado por el cliente lo guardamos en la variable "producto"
            if producto:
                cantidad_compra = int(input(f"Cantidad de {nombre_producto} que desea comprar (disponibles {producto['cantidad']}): "))
                if cantidad_compra <= producto['cantidad']:
                    total_producto = producto['precio'] * cantidad_compra
                    carrito.append({'producto': nombre_producto, 'cantidad': cantidad_compra, 'precio_total': total_producto})
                    self.actualizar_stock(nombre_producto, -cantidad_compra)
                else:
                    print(f"No hay suficiente {nombre_producto} en el inventario.")
            else:
                print(f"{nombre_producto} no existe en el inventario.")
        
        # Informacion de la compra
        total_compra = sum(item['precio_total'] for item in carrito) #"item" representa articulos en el carrito sobre los que itera el buble.
        print("Tu compra:")
        for item in carrito:
            print(f"{item['cantidad']} x {item['producto']} a {item['precio_total'] / item['cantidad']} € cada uno: {item['precio_total']} €")
        print(f"Total a pagar: {total_compra} €")
        self.ventas_totales += total_compra
        return total_compra

mi_tienda = TiendaOnline()

   
```
