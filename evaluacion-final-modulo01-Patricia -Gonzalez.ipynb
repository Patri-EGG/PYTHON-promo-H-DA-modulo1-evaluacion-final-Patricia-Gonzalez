```python
class TiendaOnline:
    def __init__(self):
        self.inventario = []  # lista vacía para almacenar productos.
        self.clientes = {}  # diccionario vacío para almacenar información del cliente.
        self.ventas_totales = 0.0 # para llevar la cuenta de ventas y que acepte numeros decimales.
    
    def agregar_producto(self, nombre, precio, cantidad): #método para añadir o actualizar productos del inventario
        for producto in self.inventario:   #Hacemos buble para iterar entre los productos del inventario.
            if producto["nombre"] == nombre: #["nombre"] se refiere al producto que se quiere agregar---el input de usuario
                producto["cantidad"] += cantidad #suma cantidad actual con cantidad actualizada
                print(f"{nombre} se ha actualizado")
                return #finaliza el método despues de actualizar.
        # si el producto no está en el inventario lo agrega.
        self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad}) #append() añade un nuevo dicionario a la lista.
        print(f"{nombre} se ha agregado al inventario") 
            
    #Metodo para visulizar el inventario actual.
    def ver_inventario(self):
        for producto in self.inventario:  # Se itera con un bucle para encontrar los productos sus valores.
            print(f'Nombre: {producto["nombre"]}, Precio: {producto["precio"]} €, Disponible: {producto["cantidad"]}')
    
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre: 
                print(f'Nombre: {producto["nombre"]}, Precio: {producto["precio"]} €, Disponible: {producto["cantidad"]}')
                return producto # Si encuentra producto lo devuelve y muestra información del  producto.
        print(f"{nombre} no está en el inventario") #Si no, solo imprime que no existe.
        # return None 
    
    def actualizar_stock(self, nombre, cantidad, precio = None): #OJO El None maneja errores de calculo en calcular inventario. 
        for producto in self.inventario:                         # Y hace que el parámetro precio sea opcional.
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad #actualiza la cantidad si el producto existe.
                return f'Inventario {nombre}  se actualizado a {producto["cantidad"]} unidades'
        self.inventario.append({"nombre": nombre,"precio": precio, "cantidad" : cantidad}) #Usamos append() al igual que en def agregar_producto
        return f'"{nombre}" se ha añadido al inventario' #Si producto no existe lo añade.
    
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                self.inventario.remove(producto) #remove() permite eliminar directamente el producto basado en su valor (en este caso nombre del producto) sin necesidad de conocer su posición en la lista.
                return f"{nombre} ha sido eliminado del inventario. "
        print(f"{nombre} no existe en el inventario. No se puede borrar.")
    
    def calcular_valor_inventario(self):
        valor_total = 0 # garantiza que la suma total comience desde cero y no desde un valor indeterminado. antes de aplicarlo tenia errors con el calculo_inventario.
        for producto in self.inventario:
         # .get() para obtener el precio si esta disponible, si no. Lo trata como predeterminado 0
            precio = producto.get("precio", 0) #Usamos .get() porque permite especificar un valor predeterminado de 0 si la clave no existe en el diccionario.
            cantidad = producto.get("cantidad", 0)
            valor_total += precio * cantidad
            return f"El valor total del inventario es {valor_total} €."
    
    def realizar_compra(self):
        print(f"¡Bienvenid@! puedes escoger entre todos nuestro productos que ves en el inventario:") 
        self.ver_inventario() #muestra inventario al usuario
        carrito = []  # variable carrito vacía para poder iniciar 
        while True:
            nombre_producto = input("Ingresa los productos que deseas comprar y Escribe ('salir') para ir al carrito").lower()
            if nombre_producto == "salir":
                break # DETIENE el bucle. Si no, el codigo continúa hasta el infinito y más allá...
            producto = self.buscar_producto(nombre_producto) #buscamos el producto ingresado por el cliente y lo guardamos en la variable "producto"
            if producto:          #uso de int() para que el input del usuario sean numericos. 
                cantidad_compra = int(input(f"Cantidad de {nombre_producto} que desea comprar (disponibles {producto['cantidad']}): "))
                if cantidad_compra <= producto['cantidad']:
                    total_producto = producto['precio'] * cantidad_compra
                    #Toda esta información la pasamos a carrito con carrito.append() 
                    carrito.append({'producto': nombre_producto, 'cantidad': cantidad_compra, 'precio_total': total_producto})
                    self.actualizar_stock(nombre_producto, -cantidad_compra) #actualizamos inventario
                else:
                    print(f"No hay suficiente {nombre_producto} en el inventario.")
            else:
                print(f"{nombre_producto} no existe en el inventario.")
        
        # Este es el carrito -Informacion de la compra 
        total_compra = sum(elemento['precio_total'] for elemento in carrito) #"elemento" representa articulos en el carrito sobre los que itera el buble.
        print("Tu compra:")
        for elemento in carrito:
            print(f"{elemento['cantidad']} x {elemento['producto']} a {elemento['precio_total'] / elemento['cantidad']} € cada uno: {elemento['precio_total']} €")
        print(f"Total a pagar: {total_compra} €")
        self.ventas_totales += total_compra
        return total_compra
```
